### Image In-Painting

```python
# pt1 = (x1,y1),pt2 = (x2,y2)
w2 = cv2.rectangle(s1,(5,5),(25,30),(255,0,0),2)
w2 = cv2.rectangle(s1,(5,35),(25,60),(0,255,0),-1)
w2 = cv2.circle(s1,(200,30),15,(0,0,255),2)    # 200,30 is the center here
w2 = cv2.circle(s1,(50,30),12,(128,128,0),-1)
w2 = cv2.putText(s1,'Kneau',(1,125),cv2.FONT_HERSHEY_TRIPLEX,1,(128,0,0),2)
plt.imshow(w2)
plt.show()
```

### Image Filter


```python
f = cv2.imread('flower.jpg')
f = cv2.cvtColor(f,cv2.COLOR_BGR2RGB)
plt.imshow(f)
plt.show()
```

#### Blurring Using the Filter


```python
k1 = np.ones((3,3))/9
k2 = np.ones((5,5))/25
k3 = np.ones((7,7))/49

r1 = cv2.filter2D(f,-1,k1)
r2 = cv2.filter2D(f,-1,k2)
r3 = cv2.filter2D(f,-1,k3)

plt.imshow(f),plt.title('Original'),plt.show()
plt.imshow(r1),plt.title('3*3 Filter'),plt.show()
plt.imshow(r2),plt.title('5*5 Filter'),plt.show()
plt.imshow(r3),plt.title('7*7 Filter'),plt.show()
```

### Sharpening and Embossing

```python
s1 = cv2.imread('kneau.jpg')
plt.imshow(s1)
plt.show()
```

```python
ker_sharp = np.array([[-1,-1,-1],[-1,9,1],[-1,-1,-1]])  # sharpening
ker_emboss = np.array([[0,-1,-1],[1,0,-1],[1,1,1]]) # embossing
```

```python
def sharp_emboss(img):
    ker_sharp = np.array([[-1,-1,-1],[-1,9,1],[-1,-1,-1]])  # sharpening
    ker_emboss = np.array([[0,-1,-1],[1,0,-1],[1,1,1]]) # embossing
    res1 = cv2.filter2D(s1,-1,ker_sharp)
    res2 = cv2.filter2D(s1,-1,ker_emboss)
    
    plt.imshow(img),plt.title('Original'),plt.show()
    plt.imshow(res1),plt.title('Sharpend Image'),plt.show()
    plt.imshow(res2),plt.title('Embossed Image'),plt.show()
```

```python
sharp_emboss(s1)
```
